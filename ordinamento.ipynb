**insertion sort**

```python
def insertion_sort(array):
    for posizione in range(1,len(array)):
        key=array[posizione]
        j=posizione-1
        while j>=0 and array[j]>key:
            bob=array[j]
            array[j]=key
            key=bob

    return array
bob=[5,3,4,1,2]
insertion_sort(bob)
```
